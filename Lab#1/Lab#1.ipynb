```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
uniform_coefficient = 1
normal_coefficient = 0.0000005
distribution_size = 10000
```

```python
def print_metrics(distribution):
    mean = np.median(distribution)
    var = np.var(distribution)
    std = np.std(distribution)

    print(f'Distribution: {distribution}')
    print(f'Mean = {mean}')
    print(f'Variation = {var}')
    print(f'Standard deviation = {std} ')
    print('=====================================')
```

```python
def uniform_constant(i, coefficient):
    uniform_distribution = np.random.rand(i)
    print('Uniform distribution:')
    print_metrics(uniform_distribution)
    plt.hist(uniform_distribution, bins=20, facecolor='red', alpha=0.5)
    plt.show()

    correction_distribution = np.zeros(i)
    final_distribution = np.zeros(i)
    for k in range(i):
        correction_distribution[k] = coefficient
        final_distribution[k] = correction_distribution[k] + uniform_distribution[k]
    print('Final distribution')
    print_metrics(final_distribution)

    plt.plot(final_distribution, label='Final')
    plt.plot(correction_distribution, label='Correction')
    plt.show()

    plt.hist(final_distribution, bins=20, alpha=0.33, facecolor='blue', label='Final')
    plt.hist(correction_distribution, bins=20, alpha=0.33, facecolor='red', label='Correction')
    plt.hist(uniform_distribution, bins=20, alpha=0.33, facecolor='green', label='Uniform')
    plt.show()

    final_without_correction = final_distribution - correction_distribution
    print('Final distribution (without correction)')
    print_metrics(final_without_correction)

    plt.hist(final_distribution, bins=20, alpha=0.5, facecolor='blue', label='Final')
    plt.hist(final_without_correction, bins=20, alpha=0.5, facecolor='yellow', label='Final(without correction)')
    plt.hist(correction_distribution, bins=20, alpha=0.5, facecolor='red', label='Correction')
    plt.hist(uniform_distribution, bins=20, alpha=0.5, facecolor='green', label='Uniform')
    plt.show()
```

```python
def normal_square(i, coefficient):
    normal_distribution = np.random.randn(i)
    print('Normal distribution:')
    print_metrics(normal_distribution)
    plt.hist(normal_distribution, bins=20, facecolor='red', alpha=0.5)
    plt.show()

    correction_distribution = np.zeros(i)
    final_distribution = np.zeros(i)
    for k in range(i):
        correction_distribution[k] = coefficient * k * k
        final_distribution[k] = correction_distribution[k] + normal_distribution[k]

    plt.plot(final_distribution, label='Final')
    plt.plot(correction_distribution, label='Correction')
    plt.show()

    plt.hist(final_distribution, bins=20, alpha=0.33, facecolor='blue', label='Final')
    plt.hist(correction_distribution, bins=20, alpha=0.33, facecolor='red', label='Correction')
    plt.hist(normal_distribution, bins=20, alpha=0.33, facecolor='green', label='Normal')
    plt.show()

    final_without_correction = final_distribution - correction_distribution
    print('Final distribution (without correction)')
    print_metrics(final_without_correction)

    plt.hist(final_distribution, bins=20, alpha=0.5, facecolor='blue', label='Final')
    plt.hist(final_without_correction, bins=20, alpha=0.5, facecolor='yellow', label='Final(without correction)')
    plt.hist(correction_distribution, bins=20, alpha=0.5, facecolor='red', label='Correction')
    plt.hist(normal_distribution, bins=20, alpha=0.5, facecolor='green', label='Normal')
    plt.show()
```

```python
uniform_constant(distribution_size, uniform_coefficient)
```

```python
normal_square(distribution_size, normal_coefficient)
```
