```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
uniform_coefficient = 1
normal_coefficient = 0.0000002
distribution_size = 10000
```

```python
def print_metrics(distribution):
    mean = np.median(distribution)
    var = np.var(distribution)
    std = np.std(distribution)

    print(f'Mean = {mean}')
    print(f'Variation = {var}')
    print(f'Standard deviation = {std} ')
```

```python
def uniform_constant(size, coefficient):
    print('Uniform distribution:')
    uniform = np.random.uniform(size=size)
    plt.hist(uniform, bins=20, facecolor='green', alpha=0.5, label='Uniform')
    plt.legend()
    plt.show()
    print_metrics(uniform)

    print('==========================')
    print('Constant distribution(constant)')
    constant = np.array([coefficient for _ in np.arange(size)])
    plt.hist(constant, bins=20, facecolor='red', alpha=0.5, label='Constant')
    plt.legend()
    plt.show()
    print_metrics(constant)
    
    print('==========================')
    print('Final distribution')
    final_distribution = np.array([c + u for c,u in zip(constant, uniform)])
    plt.plot(final_distribution, label='Final')
    plt.plot(constant, label='Real')
    plt.legend()
    plt.show()
    print_metrics(final_distribution)

    print('==========================')
    print('Comparing results')
    plt.hist(final_distribution, bins=20, alpha=0.5, facecolor='orange', label='Final')
    plt.hist(constant, bins=20, alpha=0.5, facecolor='green', label='Constant')
    plt.hist(uniform, bins=20, alpha=0.5, facecolor='red', label='Uniform')
    plt.legend()
    plt.show()
    
    print('==========================')
    print('Final distribution (without correction)')
    final_without_correction = final_distribution - constant
    plt.hist(final_distribution, bins=20, alpha=0.5, facecolor='orange', label='Final')
    plt.hist(final_without_correction, bins=20, alpha=0.5, facecolor='blue', label='Final(without correction)')
    plt.hist(constant, bins=20, alpha=0.5, facecolor='green', label='Constant')
    plt.hist(uniform, bins=20, alpha=0.5, facecolor='red', label='Uniform')
    plt.legend()
    plt.show()
    print_metrics(final_without_correction)
```

```python
uniform_constant(distribution_size, uniform_coefficient)
```

```python
def normal_square(i, coefficient):
    print('Normal distribution:')
    normal = np.random.normal(size=i)
    plt.hist(normal, bins=20, facecolor='green', alpha=0.5, label='Normal')
    plt.legend()
    plt.show()
    print_metrics(normal)

    print('==========================')
    print('Square distribution')
    square = [coefficient * (k**2) for k in np.arange(0, i ,1)]
    plt.hist(square, bins=20, facecolor='red', alpha=0.5, label='Square')
    plt.legend()
    plt.show()
    print_metrics(square)
    
    print('==========================')
    print('Final distribution')
    
    final_distribution = np.array([s + n for s,n in zip(square, normal)])
        
    plt.plot(final_distribution, label='Final')
    plt.plot(square, label='Square')
    plt.legend()
    plt.show()
    print_metrics(final_distribution)

    print('==========================')    
    print('Comparing results')
    plt.hist(final_distribution, bins=20, alpha=0.5, facecolor='orange', label='Final')
    plt.hist(square, bins=20, alpha=0.5, facecolor='green', label='Square')
    plt.hist(normal, bins=20, alpha=0.5, facecolor='red', label='Normal')
    plt.legend()
    plt.show()

    print('==========================')
    print('Final distribution (without correction)')
    # removing noise from final distribution(Monte-Carlo method)
    final_without_correction = final_distribution - square
    plt.hist(final_distribution, bins=20, alpha=0.5, facecolor='orange', label='Final')
    plt.hist(final_without_correction, bins=20, alpha=0.5, facecolor='blue', label='Final(without correction)')
    plt.hist(square, bins=20, alpha=0.5, facecolor='green', label='Square')
    plt.hist(normal, bins=20, alpha=0.5, facecolor='red', label='Normal')
    plt.legend()
    plt.show()
    print_metrics(final_without_correction)
```

```python
normal_square(distribution_size, normal_coefficient)
```
